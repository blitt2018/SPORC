```python
#top10cats = pd.DataFrame(df["category1"].value_counts()).reset_index().head(10)["category1"]
top10cats = ["religion", "society", "education", "business", "news", "sports", "comedy", "health", "arts", "music"]
```

```python
df["category1"].unique()
```

```python
#what happens if we just do tf-idf by category? 
dfList = []

for cat in top10cats: 
    tfidf = TfidfVectorizer(min_df=20, max_df = 2500, stop_words="english", max_features=1000)
    catDf = df[df["category1"] == cat]
    catFeatures = tfidf.fit_transform(list(catDf.sample(10000)["transcript"]))
    catSums = catFeatures.sum(axis=0).tolist()[0]
    names = list(tfidf.get_feature_names_out()) 
    tfidfDf = pd.DataFrame({"featureValue":catSums, "features":names})
    tfidfDf["category"] = cat 
    dfList.append(tfidfDf.sort_values("featureValue", ascending=False).head(20))
```

```python
catCatDf = pd.concat([currDf["features"].reset_index(drop=True) for currDf in dfList], axis=1)
catCatDf.columns = [top10cats]
catCatDf = catCatDf.reset_index(names="word ranking")
```
